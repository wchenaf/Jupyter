```python
import os
import pickle

def save_obj(obj, name ):
    with open('obj/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open('obj/' + name + '.pkl', 'rb') as f:
        return pickle.load(f)

def get_size(start_path = '.'):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(start_path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            total_size += os.path.getsize(fp)
    return (total_size/1024/1024)



def largets_file(start_path = '.'):
    largest=''
    max_size=0
    for f in next( os.walk(start_path) )[2]:
        fp = os.path.join(start_path, f)
        size=os.path.getsize(fp)
        if size>max_size:
            max_size=size
            largest=fp
    return largest

def largets_file_size(start_path = '.'):
    #print(start_path)
    max_size=0
    if len(next( os.walk(start_path) )[2])>0:
        for f in next( os.walk(start_path) )[2]:
            fp = os.path.join(start_path, f)
            size=os.path.getsize(fp)
            if size>max_size:
                max_size=size
    return max_size/1024/1024



namelist=[]


drivelist=[]
drivelist.append(os.walk(r"g:/Best"))
drivelist.append(os.walk(r"g:/Checked"))
drivelist.append(os.walk(r"g:/Collected"))
drivelist.append(os.walk(r"g:/DELETED"))
drivelist.append(os.walk(r"f:/VR"))
#drivelist.append(os.walk(r"g:/renamed"))
drivelist.append(os.walk(r"f:/Checked"))
drivelist.append(os.walk(r"f:/Uncensored"))
#drivelist.append(os.walk(r"d:/1/"))
#drivelist.append(os.walk(r"g:/中字种子/"))

try:
    for g in drivelist:
        print(g)
        for path,dir_list,file_list in g: 
            #print(path)
            if path.find('种子') == -1 or 1:
                for dir_name in dir_list:
                    index_=dir_name.find('-')
                    if (index_!=-1) and (index_<len(dir_name)-1):
                        #print(dir_name)
                        indexnum=index_+1
                        while (dir_name[indexnum].isdigit()) and (indexnum<len(dir_name)):
                            indexnum+=1
                            if indexnum==len(dir_name):
                                break
                        code=dir_name[:indexnum].upper()

                        #append name and path to namelist
                        if dir_name[indexnum-1]!='-':
                            #print(code)
                            path1=path+'/'+dir_name
                            namelist.append([code,[path1,str(round(get_size(path1)))+'MB']])
    namelist.sort()
    save_obj(namelist,'namelist')
    
    f = open('namelist.txt','w',encoding='utf-8')
    for x in namelist:
        #print(x)
        print(x,file=f)

    print('namelist also saved as an obj in obj/')

except BaseException:
    import sys
    print(sys.exc_info()[0])
    import traceback
    print(traceback.format_exc())
finally:
    print("Press Enter to continue ...")
    #input()

    
    
    

collectedlist=[]
for dir_name in next(os.walk('g:/Collected/'))[1]:
    collectedlist.append(dir_name)
save_obj(collectedlist,'collectedlist')


f=open('private.txt',mode='r',encoding='utf-8')
privatefile=f.read().split('\n')
privatelist=[]
for x in privatefile:
    indexny=x.find(' - 女優')
    if indexny!=-1:
        privatelist.append(x[:indexny])
save_obj(privatelist,'privatelist')


f=open('soso.txt',mode='r',encoding='utf-8')
sosofile=f.read().split('\n')
sosolist=[]
for x in sosofile:
    indexny=x.find(' - 女優')
    if indexny!=-1:
        sosolist.append(x[:indexny])
save_obj(sosolist,'sosolist')


f=open('bad.txt',mode='r',encoding='utf-8')
badfile=f.read().split('\n')
badlist=[]
for x in badfile:
    indexny=x.find(' - 女優')
    if indexny!=-1:
        badlist.append(x[:indexny])
save_obj(badlist,'badlist')
```

```python
namelist=load_obj('namelist')
seen = {}
dupes = []

for x in namelist:
    code=x[0]
    if code not in seen:
        seen[code] = x[1]
    else:
        #if get_size(seen[x[0]])>0 and get_size(x[1])>0:
        if seen[code][0].find('dup')==-1 and seen[code][0].find('无码流出')==-1 \
        and x[1][0].find('dup')==-1 and x[1][0].find("无码流出")==-1 \
        and x[1][0].find('RECYCLE')==-1:
            dupes.append([seen[code],x[1]])

save_obj(seen,'seen')
if (len(dupes)==0):
    print('No duplication')
else:
    dupes.sort()
    for x in dupes:
        print(x[0])
        print(x[1])
        print()
```
