```python
import os 
import cv2
import pickle
import datetime



#list of [片名，路径]
namelist=[]

#已下载无字幕番号合集
nonsublist=[]


def get_size(start_path = '.'):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(start_path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            total_size += os.path.getsize(fp)
    return (total_size/1024/1024)



def largets_file(start_path = '.'):
    largest=''
    max_size=0
    for f in next( os.walk(start_path) )[2]:
        fp = os.path.join(start_path, f)
        size=os.path.getsize(fp)
        if size>max_size:
            max_size=size
            largest=fp
    return largest

def largets_file_size(start_path = '.'):
    #print(start_path)
    max_size=0
    if len(next( os.walk(start_path) )[2])>0:
        for f in next( os.walk(start_path) )[2]:
            fp = os.path.join(start_path, f)
            size=os.path.getsize(fp)
            if size>max_size:
                max_size=size
    return max_size/1024/1024

def save_obj(obj, name ):
    with open('obj/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open('obj/' + name + '.pkl', 'rb') as f:
        return pickle.load(f)



def get_video_height(path):
    vid = cv2.VideoCapture(path)
    height = vid.get(cv2.CAP_PROP_FRAME_HEIGHT)
    return height

def get_size(start_path = '.'):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(start_path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            total_size += os.path.getsize(fp)
    return (total_size/1024/1024)

def largets_file(start_path = '.'):
    largest=''
    max_size=0
    for f in next( os.walk(start_path) )[2]:
        fp = os.path.join(start_path, f)
        size=os.path.getsize(fp)
        if size>max_size:
            max_size=size
            largest=fp
    return largest

                
def folder_one_level(root,minsize=0):
    for file_name in next( os.walk(root) )[2]:
        
        fp = os.path.join(root, file_name)
        size=os.path.getsize(fp)/1024/1024
        if size>minsize:
            path=root+'/'+file_name
            index=file_name.rfind('.')
            suffix=file_name[index:]
            video_name=file_name[:index]

            dst=root+'/'+video_name+'/'+file_name

            if not os.path.isdir(root+'/'+video_name):
                os.makedirs(root+'/'+video_name)
            try:
                os.rename(path,dst)
                print(path)
                print(dst)
            except Exception as e:
                print(e)
                print(' ')
    
def final_rename(dir_name,new_name,dst_dir='g:/renamed'):
    new_name.strip()
    dst=os.path.join(dst_dir, new_name)
    if dst.find(' ')!=-1:
        dst=dst.replace(' ','_')
    if dir_name!=new_name:
        print('before: '+path)
        print('after:  '+dst)
        print(' ')
    try:
        os.rename(path,dst)
    except Exception as e:
        if os.path.isdir(dst):
            os.rename(path,dst+'#')
        else:
            print(e)
            print(' ')

def final_check(dir_name,new_name,dst_dir='g:/renamed'):
    new_name.strip()
    dst=os.path.join(dst_dir, new_name)
    if dir_name!=new_name:
        new_name.strip()
        print('before: '+path)
        print('after:  '+dst)
        print(' ')

def findcode(dir_name):
    index_=dir_name.find('-')
    if (index_!=-1) and (index_<len(dir_name)-1):
        indexnum=index_+1
        while (dir_name[indexnum].isdigit()) and (indexnum<len(dir_name)):
            indexnum+=1
            if indexnum==len(dir_name):
                break
        code=dir_name[:indexnum]
    return code

def prt(msg):
    print(msg)
    print(msg,file=f)

trash_dir=r'g:/Trash'

def trash(drive):
    print('-----------------------------------------------------------')
    print('- - - - - - - - - - - - - - - - - - - - - - - - - - - - - -')
    print('NOW COLLECT TRASH UNDER '+drive)
    print('- - - - - - - - - - - - - - - - - - - - - - - - - - - - - -')
    g=next(os.walk(drive))[1]
    for dir_name in g: 
        path=drive+'/'+dir_name
        trashpath=path+'/trash'
        try:
            os.makedirs(trashpath)
        except Exception as e:
            None
        for dir_name1 in next(os.walk(path))[1]:
            if dir_name1!='trash':
                path1=path+'/'+dir_name1
                dst=trashpath+'/'+dir_name1
                os.rename(path1,dst)
        for file_name in next(os.walk(path))[2]:
            trashed=0
            indexh=dir_name.find('-')
            if indexh!=-1:
                prefix=dir_name[:indexh].lower()
                #print(prefix)
                if file_name.lower().find(prefix)==-1:
                    path1=path+'/'+file_name
                    dst=trashpath+'/'+file_name
                    #print(path1)
                    #print(dst)
                    try:
                        os.rename(path1,dst)
                        trashed=1
                    except Exception as e:
                        print(e)
                        print(' ')
                        
            if (file_name.find('.bc!')!=-1 or file_name.find('-s.jpg')!=-1)and trashed==0:
                path1=path+'/'+file_name
                dst=trashpath+'/'+file_name
                #print(path1)
                #print(dst)
                try:
                    os.rename(path1,dst)
                except Exception as e:
                    print(e)
                    print(' ')
        if len(os.listdir(trashpath))>0:
            print('trash collected in '+path+':')
            #print(os.listdir(trashpath))
            if get_size(trashpath)>(get_size(path)/3):
                print(path+' moved too much into trash')
            print()
        else:
            os.removedirs(trashpath)
    
    print('-----------------------------------------------------------')
    print('-----------------------------------------------------------')
    for dir_name in g: 
        path=drive+'/'+dir_name
        if len(os.listdir(path))>5:
            print(path+' still has more than 5 items')
```
