```python
import os 
import cv2
import pickle

def save_obj(obj, name ):
    with open('obj/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open('obj/' + name + '.pkl', 'rb') as f:
        return pickle.load(f)



def get_video_height(path):
    vid = cv2.VideoCapture(path)
    height = vid.get(cv2.CAP_PROP_FRAME_HEIGHT)
    return height

def get_size(start_path = '.'):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(start_path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            total_size += os.path.getsize(fp)
    return (total_size/1024/1024)

def largets_file(start_path = '.'):
    largest=''
    max_size=0
    for f in next( os.walk(start_path) )[2]:
        fp = os.path.join(start_path, f)
        size=os.path.getsize(fp)
        if size>max_size:
            max_size=size
            largest=fp
    return largest

def rename_one_level(root):
    print(next( os.walk(root) )[1])
    for dir_name in next( os.walk(root) )[1]:
        
        path=os.path.join(root, dir_name)
        if (get_size(path)>500) :
            video_name=largets_file(path)
            index=video_name.rfind('.')
            suffix=video_name[index:]
            dst=path+'/'+dir_name+suffix
            
            try:
                
                os.rename(video_name,dst)
                print(dst)
            except:
                print(video_name)
                
def folder_one_level(root,minsize=0):
    print(next( os.walk(root) )[2])
    for file_name in next( os.walk(root) )[2]:
        
        fp = os.path.join(root, file_name)
        size=os.path.getsize(fp)/1024/1024
        if size>minsize:
            path=root+'/'+file_name
            index=file_name.rfind('.')
            suffix=file_name[index:]
            video_name=file_name[:index]

            dst=root+'/'+video_name+'/'+file_name

            if not os.path.isdir(root+'/'+video_name):
                os.makedirs(root+'/'+video_name)
            try:


                os.rename(path,dst)
                print(path)
                print(dst)
            except:
                print(file_name)
    
def final_rename(dir_name,new_name,dst_dir='g:/renamed'):
    new_name.strip()
    dst=os.path.join(dst_dir, new_name)
    if dir_name!=new_name:
        print('before: '+path)
        print('after:  '+dst)
        print()
    os.rename(path,dst)
        
def final_check(dir_name,new_name,dst_dir='g:/renamed'):
    new_name.strip()
    dst=os.path.join(dst_dir, new_name)
    if dir_name!=new_name:
        new_name.strip()
        print('before: '+path)
        print('after:  '+dst)
        print()



```

```python
root=r'g:/1'
#print(next( os.walk(root) )[1])
prefix_list=load_obj('prefix_list')

for dir_name in next( os.walk(root) )[1]:

    path=os.path.join(root, dir_name)
    if get_size(path)>0:
        new_name=dir_name
#         print()
#         print('next dir: '+new_name)

        #去后缀
        index_dot=new_name.rfind('.')
        if index_dot>0 and index_dot==len(new_name)-4:
            #print('with dot: '+new_name)
            new_name=new_name[:index_dot]
            #print('after: '+new_name)
            
        
            
        index_c=new_name.lower().rfind('c')
        if index_c>0 and index_c==len(new_name)-1:
            if (new_name[len(new_name)-1]=='-') or (new_name[len(new_name)-1]==',') or (new_name[len(new_name)-1]=='_')\
            or (new_name[len(new_name)-1]=='.') or (new_name[len(new_name)-1]==']'):
                new_name=new_name[:len(new_name)-1]
            #print('with dot: '+new_name)
            new_name=new_name[:index_c]+'_CAVI'
            #print('after: '+new_name)
            
        index_c=new_name.rfind('_ch')
        if index_c>0 and index_c==len(new_name)-3:
            new_name=new_name[:index_c]+'_CAVI'
            #print('after: '+new_name)
            
        index_c=new_name.rfind('CMP4')
        if index_c>0 and index_c==len(new_name)-4:
            new_name=new_name[:index_c]+'_CAVI'
            #print('after: '+new_name)
            
        
            
        index_c=new_name.lower().rfind('r')
        if index_c>0 and index_c==len(new_name)-1 and new_name.find('HBR')==-1:
            if (new_name[len(new_name)-1]=='-') or (new_name[len(new_name)-1]==',') or (new_name[len(new_name)-1]=='_')\
            or (new_name[len(new_name)-1]=='.') or (new_name[len(new_name)-1]==']'):
                new_name=new_name[:len(new_name)-1]
            #print('with dot: '+new_name)
            new_name=new_name[:index_c]+'_CAVI'
            #print('after: '+new_name)    
            
        index_c=new_name.rfind('__CAVI')
        if index_c>0:
            new_name=new_name[:index_c]+'_CAVI'
            #print('after: '+new_name)
        
        
        index_c=new_name.rfind('.HD')
        if index_c>0 and index_c==len(new_name)-3:
            #print('with dot: '+new_name)
            new_name=new_name[:index_c]
            #print('after: '+new_name)
        
        index_c=new_name.rfind('CH1080P')
        if index_c>0:
            new_name=new_name[:index_c]+'_CAVI FHD'
            #print('after: '+new_name)

        
        
        #去[thz.la] [.me] [censored]
        is_thz=new_name.lower().find('thz')
        is_me=new_name.lower().find('.me')
        if is_thz!=-1 or is_me!=-1:
            index1=new_name.rfind(']')
            new_name=new_name[index1+1:]

        is_censored=new_name.lower().find('sored]')
        if is_censored!=-1:
            new_name=new_name[is_censored+6:]
            
            
        #HBR
        videopath=largets_file(path)
        indexslash=videopath.rfind('\\')
        firstthree=largets_file(path)[indexslash+1:indexslash+4].lower()
        if firstthree.lower()=='hd-' and new_name.find('HBR')==-1:
            new_name=new_name+' HBR'

            
        
        #最后去垃圾
        if (new_name[len(new_name)-1]=='-') or (new_name[len(new_name)-1]==',') or (new_name[len(new_name)-1]=='_')\
        or (new_name[len(new_name)-1]=='.') or (new_name[len(new_name)-1]==']'):
            new_name=new_name[:len(new_name)-1]
            
        final_rename(dir_name,new_name,root)
    
    
```
