# get list of genres

```python
import time
import requests
import json
import re
import pandas as pd

headers = {'User-Agent':
               'Mozilla/5.0 (Windows NT 10.0; WOW64)\
               AppleWebKit/537.36(KHTML, like Gecko)\
               Chrome/55.0.2883.87 Safari/537.36'}

codelist = []
namelist=[]


url = 'https://www.javbus.com/genre'

web = requests.get(url, headers=headers)
web.encoding = 'utl-8'

Page = web.text

#Js = json.loads(str(Page))
FFF = re.findall(r'class="col-lg-2 col-md-2 col-sm-3 col-xs-6 text-center" href=".*?<', Page)
#print(Page)

#print(FFF)
for item in FFF:
    start1 = item.rindex('/')
    end1 = item.rindex('"')
    start2 = item.rindex('>')
    end2 = item.rindex('<')
    #print(item)
    #print (item[47:end])
    codelist.append(item[start1+1:end1])
    namelist.append(item[start2+1:end2])
        

df1 = pd.DataFrame()
df1['code']=codelist
df1['name']=namelist
df1.astype('str')
df1.to_csv('list.csv', index=False, header=True)
```

# Multi-thread scrap all

```python
import time
import requests
import json
import re
import pandas as pd
from multiprocessing import Pool
from multiprocessing.dummy import Pool as ThreadPool

headers = {'User-Agent':
               'Mozilla/5.0 (Windows NT 10.0; WOW64)\
               AppleWebKit/537.36(KHTML, like Gecko)\
               Chrome/55.0.2883.87 Safari/537.36'}


def scrap_ith_genre(i):
    #print(str(i)+'  '+df1.iloc[i]['code']+' '+df1.iloc[i]['name'])
    namelist = []
    for j in range(1,1000):
        #print(str(i)+'  '+df1.iloc[i]['code']+' '+df1.iloc[i]['name'])
        #print(j)
        url = 'https://www.javbus.com/genre/'+str(df1.iloc[i]['code'])+'/'+ str(j)
        #print(url)
        web = requests.get(url, headers=headers)
        web.encoding = 'utl-8'

        Page = web.text

        #Js = json.loads(str(Page))
        FFF = re.findall(r'class="movie-box" href=".*?">', Page)
        #print(Page)

        #print(FFF)
        if len(FFF)<1 : 
            break
        for item in FFF:
            end = item.rindex('"')
            #print (item[47:end])
            namelist.append(item[47:end])
    
    
    df_new = pd.DataFrame(namelist)
    if len(df_new) > 0 :
        df_new.columns=['Name']
        df_new[df1.iloc[i]['name']]=1
    return df_new
    #df_new.astype('str')
    #print(df_new)

d = {'Name': []}
#whole_list=pd.DataFrame(d)
whole_list=pd.read_csv('out.csv',encoding = "utf-8")


df1=pd.read_csv('list.csv')
df1.astype('str')

list1=list(range(0,len(df1['name'])))
pool = ThreadPool()
results=pool.map(scrap_ith_genre,list1)
pool.close()
pool.join()

for i in range(0,len(results)):
    if (len(results[i])>0):
        whole_list = pd.merge(whole_list,results[i],how='outer')
    
whole_list.to_csv('out.csv', index=False, header=True)
```

```python
def scrap_genre(code):
    namelist = []
    for j in range(1,1000):
        print(j)
        url = 'https://www.javbus.com/genre/'+code+'/'+ str(j)
        #print(url)
        web = requests.get(url, headers=headers)
        web.encoding = 'utl-8'

        Page = web.text

        #Js = json.loads(str(Page))
        FFF = re.findall(r'class="movie-box" href=".*?">', Page)
        #print(Page)

        #print(FFF)
        if len(FFF)<1 : 
            break
        for item in FFF:
            start= item.rindex('/')
            end = item.rindex('"')
            #print (item[start+1:end])
            namelist.append(item[start+1:end])
    
    
    df_new = pd.DataFrame(namelist)
    #print(df_new)
    df_new.columns=['Name']
    df_new[code]=True
    return df_new
    #df_new.astype('str')
    #print(df_new)
    

    
whole_list=pd.read_csv('out.csv',encoding = "utf-8")
df_new=scrap_genre('2l')


whole_list = pd.merge(whole_list,df_new,how='outer')
    

whole_list.to_csv('out.csv', index=False, header=True)
```
