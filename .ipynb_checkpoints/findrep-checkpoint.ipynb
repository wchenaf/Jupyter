```python
import os 


indexnum=0

#list of [片名，路径]
namelist=[]

#已下载无字幕番号合集
nonsublist=[]


def get_size(start_path = '.'):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(start_path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            total_size += os.path.getsize(fp)
    return (total_size/1024/1024)



def largets_file(start_path = '.'):
    largest=''
    max_size=0
    for f in next( os.walk(start_path) )[2]:
        fp = os.path.join(start_path, f)
        size=os.path.getsize(fp)
        if size>max_size:
            max_size=size
            largest=fp
    return largest

def largets_file_size(start_path = '.'):
    #print(start_path)
    max_size=0
    if len(next( os.walk(start_path) )[2])>0:
        for f in next( os.walk(start_path) )[2]:
            fp = os.path.join(start_path, f)
            size=os.path.getsize(fp)
            if size>max_size:
                max_size=size
    return max_size/1024/1024
```

```python
#已下载无字幕番号合集
nonsublist=[]


drivelist=[]
drivelist.append(os.walk(r"g:/Best"))
drivelist.append(os.walk(r"g:/BEST1"))
#drivelist.append(os.walk(r"g:/Checked"))
drivelist.append(os.walk(r"g:/Collected"))

#drivelist.append(os.walk(r"i:/Checked (Non Subbed)/"))


for g in drivelist:
    for path,dir_list,file_list in g:  
        for dir_name in dir_list:
            index_=dir_name.find('-')
            if (index_!=-1) and (index_<len(dir_name)-1):
                indexnum=index_+1
                while (dir_name[indexnum].isdigit()) and (indexnum<len(dir_name)):
                    indexnum+=1
                    if indexnum==len(dir_name):
                        break
                code=dir_name[:indexnum]
                
                if (code.find('-')==code.rfind('-')) and (code.find('-')!=len(code)-1):
                    #create nonsublist
                    indexcavi=dir_name.find('CAVI')
                    if (indexcavi==-1):
                        nonsublist.append(code)

nonsublist.sort()
```

```python
#list of [片名，路径]
namelist=[]


drivelist=[]
drivelist.append(os.walk(r"g:/"))
drivelist.append(os.walk(r"h:/Checked"))
drivelist.append(os.walk(r"h:/Uncensored"))
#drivelist.append(os.walk(r"d:/1/"))



for g in drivelist:
    for path,dir_list,file_list in g:  
        if path.find('种子') == -1:
            
            for dir_name in dir_list:
                index_=dir_name.find('-')
                if (index_!=-1) and (index_<len(dir_name)-1):
                    #print(dir_name)
                    indexnum=index_+1
                    while (dir_name[indexnum].isdigit()) and (indexnum<len(dir_name)):
                        indexnum+=1
                        if indexnum==len(dir_name):
                            break
                    code=dir_name[:indexnum]

                    #append name and path to namelist
                    if dir_name[indexnum-1]!='-':
                        #print(code)
                        namelist.append([code,(path+'/'+dir_name)])

seen = {}
dupes = []

for x in namelist:
    code=x[0].lower()
    if code not in seen:
        seen[code] = x[1]
    else:
        #if get_size(seen[x[0]])>0 and get_size(x[1])>0:
        if seen[code].find('dup')==-1 and seen[code].find('无码流出')==-1 \
        and x[1].find('dup')==-1 and x[1].find("无码流出")==-1 \
        and x[1].find('RECYCLE')==-1:
            dupes.append([seen[code],x[1]])
            
if (len(dupes)==0):
    print('No duplication')
else:
    dupes.sort()
    for x in dupes:
        print(x[0])
        print(str(round(get_size(x[0])))+'MB')
        print(x[1])
        print(str(round(get_size(x[1])))+'MB')
        print()
```
